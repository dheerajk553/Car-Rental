```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=merged, x='Dissolved Oxygen (mg/L)', y='Count')
plt.title("Dissolved Oxygen vs Fish Count")
plt.xlabel("Dissolved Oxygen (mg/L)")
plt.ylabel("Fish Count")
plt.show()
```

```python
merged['Date'] = pd.to_datetime(merged['Date'])

plt.figure(figsize=(10,6))
sns.lineplot(data=merged, x='Date', y='Temperature (°C)', label='Temperature')
sns.lineplot(data=merged, x='Date', y='Count', label='Fish Count')
plt.title("Temperature and Fish Count Over Time")
plt.xlabel("Date")
plt.ylabel("Values")
plt.legend()
plt.show()
```
